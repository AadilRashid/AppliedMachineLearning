```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X, y = ml.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
fig, axes = plt.subplots(1, 3, figsize=(10, 3))
for n_neighbors, ax in zip([1, 3, 9], axes):
  # the fit method returns the object self, so we can instantiate
  # and fit in one line
  clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
  ml.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
  ml.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
  ax.set_title("{} neighbor(s)".format(n_neighbors))
  ax.set_xlabel("feature 0")
  ax.set_ylabel("feature 1")
axes[0].legend(loc=3)
```

#Now let's check how the no. of neghbors effect the accuracies

```
from sklearn.datasets import load_breast_cancer
cls = load_breast_cancer()

X_train, X_test, y_train, y_test=  train_test_split(cls.data,cls.target,stratify = cancer.target,random_state = 66)
testing_accuracy = []
training_accuracy = []
neighbors = range(1,11)

for i in neighbors:
  cls = KNeighborsClassifier(n_neighbors =  i).fit(X_train,y_train)
  training_accuracy.append(cls.score(X_train,y_train))
  testing_accuracy.append(cls.score(X_test,y_test))


plt.plot(neighbors,training_accuracy,label= "training_accuracy")
plt.plot(neighbors,testing_accuracy,label= "training_accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()



```

# **KNN has a regression brother. That is used for predicting the real value from a set of features. e.g, predicting the house price from location and size**

```
from sklearn.neighbors import KNeighborsRegressor
X,y = ml.datasets.make_wave(n_samples = 40)
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state= 0)

fig, axes = plt.subplots(1, 3, figsize=(25, 4))
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
# make predictions using 1, 3, or 9 neighbors
  reg = KNeighborsRegressor(n_neighbors=n_neighbors)
  reg.fit(X_train, y_train)
  ax.plot(line, reg.predict(line))
  ax.plot(X_train, y_train, '^', c=ml.cm2(0), markersize=8)
  ax.plot(X_test, y_test, 'v', c=ml.cm2(1), markersize=8)
  ax.set_title(
  "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
  n_neighbors, reg.score(X_train, y_train),
  reg.score(X_test, y_test)))
  ax.set_xlabel("Feature")
  ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",
"Test data/target"], loc="best")


```

# **Section 2.2: Linear models**

```
from sklearn.linear_model import LinearRegression
X,y = ml.datasets.make_wave(n_samples=42)
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42)

lr = LinearRegression().fit(X_train,y_train)

print(f"Coeffitient {lr.coef_}")
print(f"Intercept {lr.intercept_}")
```

```
print(f"Training Set Score {lr.score(X_train,y_train)}")
print(f"Testing  Set Score {lr.score(X_test,y_test)}")
```
